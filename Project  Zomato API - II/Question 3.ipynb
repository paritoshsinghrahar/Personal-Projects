### Visualization

#### Question 1

#### Plot the bar graph top 15 restaurants have a maximum number of outlets.


```python
#setpath
path = 'zomato2.csv'

#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib 

df = pd.read_csv(path, encoding='utf-8')

ans = df['Restaurant Name'].value_counts()
index = ans.index
value = ans.values
for i in range(15):
    print(index[i],value[i])
    
x=index[:15]
y=value[:15]

#Plot Graph
plt.title('No. of Restaurants', fontsize="15")
plt.xlabel('Restaurant Name',fontsize="12")
plt.ylabel('No. of Restaurant',fontsize="12")
plt.xticks(rotation = 90)
plt.grid()
matplotlib.rcParams['figure.figsize'] = [10,10]
plt.bar(x,y,color = 'yellow', edgecolor = 'black')
plt.show()
```

#### Question 2

#### Plot the histogram of aggregate rating of restaurant( drop the unrated restaurant).

```python
#setpath
path = 'zomato2.csv'

#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib 

df = pd.read_csv(path, encoding='utf-8')
df['Aggregate rating'].dropna(inplace=True)
m= df['Aggregate rating'].to_list()


#Plot Graph
plt.title('Histogram on Aggregate rating', fontsize="15")
plt.xlabel('Aggregate rating',fontsize="12")
plt.ylabel('Count',fontsize="12")
plt.grid()
matplotlib.rcParams['figure.figsize'] = [10,10]

plt.hist(m,bins=25)
plt.show()
```

#### Question 3

#### Plot the bar graph top 10 restaurants in the data with the highest number of votes.


```python
#setpath
path = 'zomato2.csv'

#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib 

df = pd.read_csv(path, encoding='utf-8')

df.sort_values(by=['Votes'],inplace=True,ascending=False)

x=[]
y=[]
for i in range(10):
    print(df['Restaurant Name'].iloc[i],df['Votes'].iloc[i])
    x.append(df['Restaurant Name'].iloc[i])
    y.append(df['Votes'].iloc[i])

#Plot Graph
plt.title('No. of Votes', fontsize="15")
plt.xlabel('Restaurant Name',fontsize="12")
plt.ylabel('No. of Votes',fontsize="12")
plt.xticks(rotation = 90)
plt.grid()
matplotlib.rcParams['figure.figsize'] = [10,10]
plt.bar(x,y,color = 'yellow', edgecolor = 'black')
plt.show()
```

#### Question 4

#### Plot the pie graph of top 10 cuisines present in restaurants in the USA.


```python
#setpath
path = 'zomato2.csv'

#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib 

df = pd.read_csv(path, encoding='utf-8')

df = df[df["Country Code"] == 216]

city={}

df['Cuisines'].dropna(inplace=True)

#convert function
def convertInvestorsMultiple(name):
    l = [x.strip() for x in name.split(',')]
    for i in l:
        if i != '':
            city[i] = city.get(i,0) + 1
            
df['Cuisines'].apply(convertInvestorsMultiple)

x=[]
y=[]


sorted_x = sorted(city.items(), key=lambda kv: kv[1], reverse=True)
for i in range(10):
    print(sorted_x[i][0],sorted_x[i][1])
    x.append(sorted_x[i][0])
    y.append(sorted_x[i][1])
    
#Plot Graph
plt.title('Top Cuisines of USA', fontsize="15")
plt.xlabel('Cuisines',fontsize="12")
plt.ylabel('No. of resturants',fontsize="12")
plt.xticks(rotation = 90)
plt.grid()
matplotlib.rcParams['figure.figsize'] = [10,10]
plt.bar(x,y,color = 'yellow', edgecolor = 'black')
plt.show()
```

#### Question 5


#### Plot the bubble graph of a number of Restaurants present in the city of India and keeping the weighted restaurant rating of the city in a bubble.

```python
#setpath
path = 'zomato2.csv'

#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib 

df = pd.read_csv(path, encoding='utf-8')
df['Aggregate rating'].dropna(inplace=True)
df = df[df["Country Code"] == 1]

ans = df['City'].value_counts()
index = ans.index
value = ans.values

city=[]
for i in range(len(index)):
    m = df[df['City'] == index[i]]
    
    z = m["Votes"] * m["Aggregate rating"]
    ag1 = z.sum()
    ag2 = m['Votes'].sum()
    ag3 = ag1/ag2
    city.append([index[i],value[i],ag3])

x=[]
y=[]
z=[]

for i in range(len(city)):
    #print(city[i][0],city[i][1], format(round(city[i][2], 2)) )
    x.append(city[i][0])
    y.append(city[i][1])
    z.append(city[i][2])

#Plot Graph
plt.title('Weighted restaurant rating of the city in a bubble.', fontsize="15")
plt.xlabel('City',fontsize="12")
plt.ylabel('No. of resturants',fontsize="12")
plt.xticks(rotation = 90)
matplotlib.rcParams['figure.figsize'] = [10,10]
plt.scatter(x,y,s=z,color = 'yellow', edgecolor = 'black')
plt.show()
```
