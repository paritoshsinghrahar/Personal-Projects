#### Restaurant serving more number of cuisines

```python
#setpath
path = 'zomato2.csv'

#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib 

df = pd.read_csv(path, encoding='latin-1')
df.dropna(subset=['Aggregate rating', 'Cuisines'],inplace=True)


#convert function
def convertCuisines(name):
    count=0
    l = [x.strip() for x in name.split(',')]
    for i in l:
        if i != '':
            count+=1
    return count
            
df['countscuisine'] = df['Cuisines'].apply(convertCuisines)


x=df['Aggregate rating'].to_list()
y=df['countscuisine'].to_list()

#Plot Graph    
plt.title('Cities where the most no. of funding is done.', fontsize="15")
plt.xlabel('Avg. Ratinng',fontsize="12")
plt.ylabel('Distinct cuisine',fontsize="12")

plt.title("Avg. Rating v/s Distinct cuisine")
plt.scatter(x, y)
plt.show()
```

#### Find the weighted restaurant rating of each locality and find out the top 10 localities with more weighted restaurant rating?

        Weighted Restaurant Rating=Σ (number of votes * rating) / Σ (number of votes) .

```python
#setpath
path = 'zomato2.csv'

#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib 

df = pd.read_csv(path, encoding='latin-1')
df['Aggregate rating'].dropna(inplace=True)
df['Votes'].dropna(inplace=True)


ans = df['Locality'].value_counts()
index = ans.index
value = ans.values

city=[]
for i in range(len(index)):
    m = df[df['Locality'] == index[i]]
    
    z = m["Votes"] * m["Aggregate rating"]
    ag1 = z.sum()
    ag2 = m['Votes'].sum()
    if ag1 !=0:
        ag3 = ag2/ag1
        city.append([index[i],value[i],ag3])

x=[]
y=[]

sorted_x = sorted(city, key=lambda kv: kv[2], reverse=True)
for i in range(10):
    x.append(city[i][0])
    y.append(city[i][2])


#Plot Graph
plt.title('Weighted restaurant rating of the localities', fontsize="15")
plt.xlabel('Locality',fontsize="12")
plt.ylabel('Weighted restaurant rating ',fontsize="12")
plt.xticks(rotation = 90)
matplotlib.rcParams['figure.figsize'] = [10,10]
plt.bar(x,y,color = 'yellow', edgecolor = 'black')
plt.show()
```
